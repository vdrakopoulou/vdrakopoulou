## Standardize Column Order and Concatenate DataFrames

### Subtask:
Standardize the column order for `reports_df` and `sentences_df`, then concatenate them into a single DataFrame.

#### Instructions
1. Define a list of common columns that both `reports_df` and `sentences_df` should have, in the desired order.
2. Reindex both `reports_df` and `sentences_df` to match this common column order.
3. Concatenate the two DataFrames vertically (row-wise) into a new DataFrame called `esg_messages_df`.
4. Display the first few rows of the combined `esg_messages_df`.

**Reasoning**:
The subtask requires standardizing the column order for both DataFrames and then concatenating them. This code defines a common set of columns, reindexes both `reports_df` and `sentences_df` to match this order, and then concatenates them into a single `esg_messages_df`.



```python
common_columns = [
    'company_name',
    'source_url',
    'raw_text',
    'source_type',
    'datetime',
    'ticker',
    'language',
    'verification_level',
    'engagement_count'
]

reports_df = reports_df[common_columns]
sentences_df = sentences_df[common_columns]

esg_messages_df = pd.concat([reports_df, sentences_df], ignore_index=True)

display(esg_messages_df.head())
display(esg_messages_df.tail())
```
